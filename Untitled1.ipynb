```python
from ultralytics import YOLO
model = YOLO("yolov8n.pt")
```

```python
model.train(
    data="/content/Accident-Detection-2/data.yaml",
    epochs=100,
    imgsz=640,
    batch=16,
    optimizer="AdamW",
    lr0=0.01,
    lrf=0.01,
    patience=20,
    augment=True,
    fliplr=0.5,
    mosaic=1.0,
    amp=True  # Mixed precision training
)
```

```python
# Inference on a sample image
results = model("/content/car-damage-accident-at-malshej-ghat-maharashtra-india-asia-ET0Y8C.jpg", conf=0.4)
results[0].show()
```
